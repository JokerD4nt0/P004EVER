Coder une fonction `linear_regression(theta, x)` qui retoure $y = \sum_{i=0}^n \theta_i x_i$. On prendra $x_0 = 1$

```python
import numpy as np
def linear_regression(x,theta) : # bien mettre la matrice AVANT le vecteur sinon erreur
    if len(theta)==x.shape[1]+1:
        vect_init=np.ones((len(x),1))
        x=np.concatenate((vect_init,x),axis=1)
    result=np.matmul(x,theta)
    return result
X=np.array([[1,1,1,1]])
print(X)
theta=np.array([2,4,6,2,2])
linear_regression(X,theta)
```

```python
def linear_regression_cost(theta,X,Y):
    '''
    page help
    '''
    #p('X',X)
    #p('theta',theta)
    h=linear_regression(X,theta)
    #p('h',h)
    result=1/(2*len(X))*sum((h-Y)**2)
    return result
            
X=np.array([[1,1,1,1],[2,2,2,2]])
theta = np.array([2,2,2,2,2])
Y=[3,3]
linear_regression_cost(theta,X,Y)
```

formule : J(theta) = - (ylog(h) +(1-y)log(1-h))

```python
def logistic_regression_cost(model,X,Y,theta):
    h=model(X,theta)
    #p('h',h)
    result=-(Y*np.log(h)+(1-Y)*np.log(1-h))
    return result
            
logistic_regression_cost(linear_regression,X,Y,theta)     
```
